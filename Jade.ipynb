```python
import time
import random

# Agent 1: Prediction Agent
class PredictionAgent:
    def predict_surplus(self):
        surplus = round(random.uniform(2.0, 5.0), 2)
        print(f"[PredictionAgent] Predicted Surplus: {surplus} kWh")
        return surplus

# Agent 2: Negotiation Agent
class NegotiationAgent:
    def find_trade(self, surplus):
        if surplus > 3.0:
            neighbor = "Neighbor_A"
            print(f"[NegotiationAgent] Found trading partner: {neighbor}")
            return neighbor
        else:
            print("[NegotiationAgent] No trading needed.")
            return None

# Agent 3: Smart Contract Agent
class SmartContractAgent:
    def execute_trade(self, surplus, neighbor):
        print(f"[SmartContractAgent] Executing trade: {surplus} kWh to {neighbor}")
        # Simulate blockchain delay
        time.sleep(1)
        print("[SmartContractAgent] Trade recorded on blockchain.")

# Agent 4: Facilitating Agent
class FacilitatingAgent:
    def __init__(self):
        self.prediction_agent = PredictionAgent()
        self.negotiation_agent = NegotiationAgent()
        self.contract_agent = SmartContractAgent()

    def manage_energy(self):
        print("\n[FacilitatingAgent] Starting Surplus Management Process")
        surplus = self.prediction_agent.predict_surplus()
        neighbor = self.negotiation_agent.find_trade(surplus)
        if neighbor:
            self.contract_agent.execute_trade(surplus, neighbor)
        else:
            print("[FacilitatingAgent] No trade executed.\n")

# Run the System
if __name__ == "__main__":
    facilitator = FacilitatingAgent()
    for _ in range(3):  # Simulate 3 cycles
        facilitator.manage_energy()
        time.sleep(2)
```
